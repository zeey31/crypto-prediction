```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
ticker = []
coins = {"Bitcoin" : "BTC", "Etherum" : "ETH", "Tether" : "USDT", "Usd coin" : "USDC", "Binance coin" : "BNB", "Cardano" : "ADA", "Xrp" : "XRP", "Binance usd" : "BUSD", "Solana" : "SOL", "Doge coin" : "DOGE", "Polkadot" : "DOT", "Wrapped bitcoin" : "WBTC", "Dai" : "DAI", "Litecoin" : "LTC", "Avalanche" : "AVAX", "Uniswap" : "UNI", "True usd" : "TUSD", "Filecoin" : "FIL", "Monero" : "XMR", "Tron" : "TRX" }
for key, value in coins.items():
   ticker.append(value+"-USD")


data = yf.download(ticker, start='2016-01-01')
```
